```python
cpfs=list(df.cpf.unique())
```

```python
def ult_3_qtd(x):
    if x['mês'] <=5 & x['mês']>3:
        r= 1
    else:
        r= 0
    return r

def ult_3_vlr(x):
    if x['ult_3_qtd'] ==1:
        r= x['vlr']
    else:
        r= 0
    return r

def ult_0_qtd(x):
    if x['mês'] ==5:
        r= 1
    else:
        r= 0
    return r

def ult_0_vlr(x):
    if x['ult_0_qtd'] ==1:
        r= x['vlr']
    else:
        r= 0
    return r
```

```python
lista=[]
for i in cpfs:
    base_aux=df[df.cpf==i]
    d=base_aux[['mês','cpf']].groupby(by='mês').count().reset_index()
    e=base_aux[['mês','vlr']].groupby(by='mês').sum().reset_index()
    f=pd.merge(d,e,how='left')

    f['ult_3_qtd'] = f.apply(ult_3_qtd,axis=1)
    f['ult_3_vlr'] = f.apply(ult_3_vlr,axis=1)

    f['ult_0_qtd'] = f.apply(ult_0_qtd,axis=1)
    f['ult_0_vlr'] = f.apply(ult_0_vlr,axis=1)

    
    a1=f['mês'].count()
    a2=f['cpf'].sum()
    a3=f['vlr'].sum()
    a4=f['ult_3_qtd'].sum()
    a5=f['ult_3_vlr'].sum()
    a6=f['ult_0_qtd'].sum()
    a7=f['ult_0_vlr'].sum()

    fim = pd.DataFrame({'cpf': [i],
                        'qtd_mes_total': [a1],
                        'qtd_total':[a2],
                        'vlr_total':[a3],
                        'ult_3_qtd':[a4],
                        'ult_3_vlr':[a5],
                        'ult_0_qtd':[a6],
                        'ult_0_vlr':[a7]})

    lista.append(fim)
```
